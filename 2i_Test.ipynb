# In a language of your choice could you please provide code that iterates in multiples of 7 up until 100, then in multiples of 8 up to 200, then multiples of 9 up to 300.

```python
import numpy as np 
import pandas as pd 
```

A while loop will be created to iterate by multiples of 7

```python
count = 0 # Create an iterator
Multiples_of_7 = [] # A empy list to store results 
while count < 100:# Set loop to stop when the iterator is larger than 100
    Multiples_of_7.append(count)# Add the iterators value to the list
    count+=7# increase by a factor of 7
    

print(Multiples_of_7)
```

#### This loop will be converted into a fuction in order to change the factor of iteration (7, 8, 9).
Where the factor of iteration will be called magnitude.

And the finish number (100,200,300) will be called limit.

```python
def Iterator(magnitude, limit):
    counter = 0 # start at 0
    Multiples = [] # Value Container
    while counter < limit: # Set condtion continue while the counter is lower than the limit       
        Multiples.append(counter)# Populate container
        counter+=magnitude# increase by selected magnitude
    return Multiples # return multiples of  magnitude within the limit
        
    
```

```python
M7 = Iterator(7,100) # returns the multiples of 7 within 100 in a list
M7
```

```python
M8 = Iterator(8,200) # returns the multiples of 8 within 200 in a list
M8
```

```python
M9 =  Iterator(9,300) # returns the multiples of 9 within 300 in a list
M9
```

#### The same fuction will be altered to iterate through a Dataset.

The function will return the values within the dataset which index is a multiple of the magnitude, as a pandas Dataframe for better visualization.

```python
Dataset = np.random.randint(1,101,300) # Dataset of 300 random elements from 1 to 100
Dataset
```

```python
def Data_Iterator(magnitude, limit,data):
    counter = 0
    Sub_set = []
    Index = []
    while counter < limit:
        Sub_set.append(data[counter])
        Index.append(counter)
        counter+=magnitude
    df1 = pd.DataFrame(Index,columns=[('Index')])
    df2 = pd.DataFrame(Sub_set,columns=[('Index Factor '+str(magnitude))])
    df1['Values']= df2
    return df1
        
    
```

```python
subset_7 = Data_Iterator(7,100,Dataset) # Returns the elements which index is a multiple of 7 and is within 100
subset_7
```

```python
subset_8 = Data_Iterator(8,200,Dataset)# Returns the elements which index is a multiple of 8 and is within 200
subset_8
```

```python
subset_9 = Data_Iterator(9,300,Dataset)# Returns the elements which index is a multiple of 9 and is within 300
subset_9
```
